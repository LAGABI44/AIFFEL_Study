- 터미널을 켠 직후 현재 위치를 출력하면, /content이라는 디렉토리에 위치

###(3) 현재 디렉토리 내에 있는 모든 파일 목록을 보여줘!

```python
!ls
!ls -al
```

- ls 는 List의 줄임말로, 현재 디렉토리 내의 모든 파일 또는 하위 디렉토리의 목록을 모두 출력   
- 기본 명령어 뒤에 - (dash) 하나를 입력 후 추가 명령어를 입력하면 세부 옵션을 설정할 수 있음   
- \- 뒤의 a 는 숨김 파일까지 모두 출력하라는 옵션   
-  l 은 자세히(long format) 정보를 출력하라는 의미
- 터미널에서 숨김파일은 파일/폴더명 앞에 . 이 붙게 되므로, 이전에는 나오지 않았던 여러 목록들까지 출력이 된 것을 확인할 수 있음   
- [리눅스 ls -l 명령어 - 파일 정보 확인과 의미](https://www.leafcats.com/137)   

###(4) 원하는 디렉토리로 이동해보자! 더블클릭!



```python
! cd sample_data
! pwd
```

- cd 는 Change Directory, 즉 디렉토리 변경   
- cd 명령어 뒤에 이동하기를 원하는 디렉토리 이름을 입력하면 그 디렉토리로 이동   

```python
! cd ..
! pwd
```

- cd 뒤에 원하는 디렉토리의 이름 대신, .. 을 입력하면 상위 폴더로 나가게 됨

```python
! cd ~
! pwd
```

- cd ~ 라는 명령어는 ~ 위치로 이동시키라는 명령어   
- 현재 어떤 폴더 위치에 있었든지 항상 /content 위치로 이동   
- 한 번에 Home 디렉토리로 돌아오고 싶을 때 유용한 명령어

###(5) 새 폴더 만들기, 또는 폴더 삭제하기


```python
# 폴더 만들기
! mkdir new_folder
! ls
```

- mkdir 은 Make Directory

```python
# 폴더 삭제하기
! rm -r new_folder
! ls
```

- rm -r 명령어 뒤에 삭제하길 원하는 디렉토리 명을 입력하면 삭제됨   
- rm은 Remove의 줄임말   
- -r 옵션은 Recursive라는 의미로, 디렉토리 내부의 모든 파일 및 폴더에 대해 재귀적(반복적)으로 명령을 수행하라는 의미를 가짐   
- -r 옵션은 디렉토리를 삭제할 때 필요, 하위 디렉토리와 파일까지 모두 삭제하라는 것을 의미   
- 디렉토리가 아닌 일반 파일을 삭제할 때에는 rm 명령어만 입력

###(6) 파일을 원하는 곳으로 옮기거나, 복사하기


```python
! cd ~
! mkdir new_folder
! mv new_folder content
```

- new_folder 디렉토리를 새로 만들어준 후   
- mv 명령어를 이용해서 new_folder 디렉토리를 content 안으로 옮김   
- mv 명령어는 영어 단어 Move를 줄여서 만듦   
- mv 명령어 뒤에 이동하고 싶은 파일, 이동할 목적지 디렉토리를 순서대로 입력

```python
# 잘 옮겨졌는지 확인
! cd content
! ls
```

- ls 를 입력해서 new_folder 가 목록에 나온다면 성공!

```python
# 파일 또는 디렉토리 복사하기
! cp -r new_folder ..
! cd ..
! ls
```

- new_folder 를 .. 의 위치, 즉 상위 폴더에 복사   
- cp 는 Copy의 줄임말   
- cp 명령어는 rm(삭제) 명령어와 같이 디렉토리를 복사할 때 -r 옵션을 추가해주어야 복사하려는 디렉토리의 하위디렉토리까지 함께 복사한다는 점   
- 별 파일을 복사하고 싶을 때에는 -r 이 필요없음

##패키지 설치 및 관리
###(1) 패키지란?
- 특정 기능을 하는 작은 프로그램 단위, 라이브러리와 유사   
- 패키지는 우리가 다운 받아 사용할 수 있는 다양한 툴을 제공   
- 우분투에서 패키지를 관리하기 위해 주로 쓰이는 명령어는 apt-get 
- [Ubuntu apt-get 명령어 사용법](https://codechacha.com/ko/linux-apt-commands/)

###(2) 패키지 설치

```python
#지금까지 설치된 패키지의 리스트 확인
! sudo apt list --installed
```

- sudo apt list --installed 명령어를 입력하면 설치된 패키지의 목록이 차례로 출력   
- sudo apt list --installed 명령어 뒤에 | grep (패키지명) 을 입력해주면 검색하는 단어를 포함하는 패키지만 출력
